```python
import pandas as pd
import datetime
import numpy as np
food_df = pd.read_csv('Food_Inspections.csv')
sidewalk = pd.read_csv('Sidewalk_Cafe_Permits_-_Current.csv')
```

```python
food_df.head()
```

```python
food_df.columns
```

```python
food_col = food_df[['Inspection ID', 'DBA Name','License #', 'Facility Type',
       'Risk', 'Address','Zip', 'Inspection Date',
       'Inspection Type', 'Results', 'Violations','Location' ]]

food_col
```

```python
food_na = food_col.dropna(subset=['Location'])
food_na
```

```python
sidewalk.columns
```

```python
sidewalk_col = sidewalk[['PERMIT NUMBER','DOING BUSINESS AS NAME', 'ISSUED DATE', 'EXPIRATION DATE','ADDRESS',
                        'ZIP CODE','Location']]
```

```python
sidewalk_col.head()
```

```python
sidewalk_rename = sidewalk_col.rename(columns={"LOCATION": "Location", 'ADDRESS':'Address', 'DOING BUSINESS AS NAME': 'DBA Name'})
sidewalk_rename.head()
```

```python
sidewalk_na = sidewalk_rename.dropna(subset=['Location'])
sidewalk_na
```

```python
Join = sidewalk_na.merge(food_na,on= "Location', sort=False)
```
